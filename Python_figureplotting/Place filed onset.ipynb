```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
ca1_fn_dist_fp=r'D:\DATA\analysis\15lapstandarded\delay_lap summary\CA1\CA1_onset_dist.mat'
ca3_fn_dist_fp=r'D:\DATA\analysis\15lapstandarded\delay_lap summary\CA3\CA3_onset_dist.mat'
CA1_fn_dist_data = scipy.io.loadmat(ca1_fn_dist_fp)
CA3_fn_dist_data = scipy.io.loadmat(ca3_fn_dist_fp)
```

```python
CA1_fn_dist_data
distbins=np.arange(1,31,1)
distbins
CA1_fn_dist_data['f_distribution']
```

```python
plt.figure(figsize=(20,10))
ax = sns.distplot(CA1_fn_dist_data['f_PF_start'][0],kde=False,bins=distbins,norm_hist=True,label='CA1',color='red',
                  hist_kws={"edgecolor": "k", "linewidth": 2})
ax = sns.distplot(CA3_fn_dist_data['f_PF_start'][0],kde=False,bins=distbins,norm_hist=True,label='CA3',color='black',
                   hist_kws={"edgecolor": "k", "linewidth": 2})
ax.legend()
ax.set_title('CA1-CA3_f-PF_onset')
ax.set_xlabel('onset lap')
plt.savefig('CA1-CA3_f-PF_onset.svg')
```

```python
plt.figure(figsize=(20,10))
ax = sns.distplot(CA1_fn_dist_data['n_PF_start'][0],kde=False,bins=distbins,norm_hist=True,label='CA1',color='red',
                  hist_kws={"edgecolor": "k", "linewidth": 2})
ax = sns.distplot(CA3_fn_dist_data['n_PF_start'][0],kde=False,bins=distbins,norm_hist=True,label='CA3',color='black',
                   hist_kws={"edgecolor": "k", "linewidth": 2})
ax.legend()
ax.set_title('CA1-CA3_n-PF_onset')
ax.set_xlabel('onset lap')
ax.set_xlim(0, 30)
plt.savefig('CA1-CA3_n-PF_onset.svg')
```
